### df.sort_value()

```python
import pandas as pd 

pd.set_option('display.unicode.east_asian_width', True)

df = pd.read_excel('data/电脑配件销售记录.xlsx')
print(df.head(), '\n')

# 根据 成交金额 排序
df1 = df.sort_values(by='成交金额', ascending=False)   # ascending=False 降序
print(df1.head(), '\n')

# 根据多列进行排序， 数量和成交金额
df2 = df.sort_values(by=['数量', '成交金额'], ascending=False)
print(df2, '\n')
```

### df.rank()

```python
import pandas as pd 

pd.set_option('display.unicode.east_asian_width', True)

df = pd.read_excel('data/电脑配件销售记录.xlsx')
# print(df.head(), '\n')

df = df.sort_values(by='数量', ascending=False) 
# print(df)

df['平均数量排名']=df['数量'].rank(ascending=False)  # 默认采用了顺序排名的平均值
print(df)
```

```python
import pandas as pd 

pd.set_option('display.unicode.east_asian_width', True)

df = pd.read_excel('data/电脑配件销售记录.xlsx')
# print(df.head(), '\n')

df = df.sort_values(by='数量', ascending=False) 
# print(df)

df['平均数量排名']=df['数量'].rank(method='min', ascending=False)  # 采用最小值排名
print(df)
```
