```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
x = [1, 2, 3, 4, 5]
y = [90, 45, 32, 68, 92]

# List of styles to demonstrate
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

# Create a figure with subplots for each style
plt.figure(figsize=(15, 10))

for i, style in enumerate(styles):
    sns.set_style(style)
    plt.subplot(2, 3, i + 1)
    sns.barplot(x=x, y=y)
    plt.title(style)

plt.tight_layout()
plt.show()
```

### Seaborn绘制常用的图表函数

1. 折线图

    relplot(x, y, kind, data)
    
    lineplot(x, y, data)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('data/成绩表.xlsx')

sns.relplot(x='姓名', y='数学', kind='line', data = df)
plt.xticks(df['姓名'], rotation=45)
plt.title('数学成绩')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('data/成绩表.xlsx')

sns.lineplot(x='姓名', y='数学', data = df)
plt.xticks(df['姓名'], rotation=45)
plt.title('数学成绩')

plt.show()
```

```python
#  多折线 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('data/成绩表.xlsx')
df1 = df[['姓名', '数学', '语文', '英语']]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df1.set_index('姓名'))
plt.xticks(df['姓名'], rotation=45)
plt.title('成绩表')

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = pd.read_excel('data/data2.xls')
data = df[['得分']]

sns.histplot(data, kde=True)
plt.show()
```

4. 散点图

    relplot(x, y, data)

```python
# import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('tips')
sns.relplot(x="total_bill", y="tip", data=df, color='r')
plt.show()
```

5. 线性回归模型

    lmplot(x, y, data)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('tips')

sns.lmplot(x="total_bill", y="tip", data=df)
plt.show()
```

6. 箱型图

    boxplot(data, rug)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('tips')

sns.boxplot(x="day", y="total_bill", hue='time', data=df)
plt.show()
```

7. 核密度图

    kdeplot(data, shade=True)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('iris')
# df
sns.kdeplot(df['petal_width'], shade=True, color='r', label='Petal Width')
sns.kdeplot(df['petal_length'], shade=True, color='b', label='Petal Length')
plt.legend()
plt.show()
```

```python
#  边际核密度图
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('iris')

sns.jointplot(x=df['sepal_length'], y=df['sepal_width'], kind='kde', space=0)
plt.show()
```

8. 提琴图

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.rcParams['font.sans-serif'] = ['SimHei']

df = sns.load_dataset('tips')

sns.violinplot(x='total_bill', y='day', hue='time', data=df)
plt.show()
```
