### 线条颜色和样式

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, color='green', linewidth=2, linestyle='--')
# color: 颜色, 数值： 灰度
# linewidth: 线宽
# litestyle: '--' 双划线，'-' 单划线，':' 点划线，'' 无划线


y1 = [10, 13, 16, 19, 22]
plt.plot(x, y1, color='red', linewidth=1, linestyle='-')

y2 = [37, 24, 2, 43, 54]
plt.plot(x, y2, color='0.4', linewidth=3, linestyle=':')

y1 = [1.5, 3.5, 5.7, 7.2, 9.1]
plt.plot(x, y1, color='green', linewidth=2, linestyle='')

plt.show()
```

### 文本标签_标题_图例

```python
# 设置坐标轴的刻度

import matplotlib.pyplot as plt
import random 

x = [i for i in range(1, 11)]    # 列表生成式
y = [random.randint(20, 50) for i in range(1, 11)]

plt.figure( figsize=(10, 6))
plt.plot(x, y, marker='o', mfc='w')

month = [str(i)+'月' for i in range(1, 11)]
plt.xticks(range(1, 11), month)
plt.yticks(range(20, 51))
# plt.xlim(1, 20) # 限制 x 轴的取值范围
# plt.ylim(20, 50) # 限制 y 轴的取值范围

# plt.grid(color='r', linestyle='--', axis='x') # 显示网格线, axis='x' 表示对 x 轴进行网格线
plt.grid(color='0.5', linestyle='--', axis='y') # 显示网格线, axis='y' 表示对 y 轴进行网格线

# 添加文本标签
for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=15, color='r')
    # a 和 b 分别表示 x 和 y 的坐标
    # b 表示文本内容
    # ha: horizontal alignment
    # va: vertical alignment

# 添加图标的标题
plt.title('测试练习折线图', fontsize=20)

# 添加图例
plt.legend(('销售次数',))
plt.show()
```

### 文本注释

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [10, 45, 35, 46, 7, 34]

plt.plot(x, y, marker='o')
for a,b in zip(x,y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize = 12, color='r')
# 添加文本注释
plt.annotate('最大数', xy=(4, 46), xytext=(5, 46), arrowprops=dict(facecolor='r', shrink=0.05))
# xy  被注释点的坐标。
# xytext 注释文本的坐标。
# arrowprops 注释箭头的属性。
# shrink 设置箭头相对于注释文本的缩放比例。

plt.show()
```

### 坐标刻度线

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [10, 45, 35, 46, 7, 34]

plt.rcParams['font.sans-serif'] = ['SimHei']

plt.plot(x, y, marker='o')
for a,b in zip(x,y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize = 12, color='r')

# 添加图标的标题
plt.title('测试练习折线图', fontsize=16)

# 添加图例
plt.legend(('销售次数', ), loc='lower left')

# 添加文本注释
plt.annotate('最大数', xy=(4, 46), xytext=(5, 46), arrowprops=dict(facecolor='r', shrink=0.05))
# xy  被注释点的坐标。
# xytext 注释文本的坐标。
# arrowprops 注释箭头的属性。
# shrink 设置箭头相对于注释文本的缩放比例。

# 设置间距
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [10, 45, 35, 46, 7, 34]

plt.rcParams['font.sans-serif'] = ['SimHei']

plt.plot(x, y, marker='o')
for a,b in zip(x,y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize = 12, color='r')

# 设置网格
plt.grid(color='0.5', linestyle='--')

# 添加图标的标题
plt.title('测试练习折线图', fontsize=16)

# 添加图例
plt.legend(('销售次数', ), loc='lower left')

# 添加文本注释
plt.annotate('最大数', xy=(4, 46), xytext=(5, 46), arrowprops=dict(facecolor='r', shrink=0.05))
# xy  被注释点的坐标。
# xytext 注释文本的坐标。
# arrowprops 注释箭头的属性。
# shrink 设置箭头相对于注释文本的缩放比例。

# 设置间距
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

# 设置刻度
plt.tick_params(bottom=True, left=True, right=False, top=False)

# 设置坐标轴刻度线显示的方向
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'out'

plt.xlabel('x 轴')
plt.ylabel('y 轴')

plt.show()
```
